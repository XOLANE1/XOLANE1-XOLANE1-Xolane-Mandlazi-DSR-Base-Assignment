### What is the total number of people living with HIV (NoPLHIV) in the listed districts according to the Survey estimate?

```python
#total number of people living with HIV in the listed districts according to the Survey estimate?
data_set['NoPLHIV'].where(data_set['Estimate']=='Survey').sum() 
```

### What is the average NoPLHIV of the two estimates used for “Xhariep”?

```python
#average NoPLHIV of the two estimates used for “Xhariep”
data_set['NoPLHIV'].where(data_set['District']=='Xhariep').groupby(data_set['Estimate']).mean()
```

### Add a column and populate it with the number of people not living with HIV for each row.


```python
# calculation the population of people linving with HIV
def PLHIV(Prevalence,NoPLHIV):
    No_Prevalence=[]
    TOTAL_PERC=100
    for i,j in zip(Prevalence,NoPLHIV):
        total_population=(j/(i/TOTAL_PERC))
        No_Prevalence.append(total_population-j)
    return pd.DataFrame(No_Prevalence) #convering the list into a DataFrame
```

```python
# Passsing the neccessary columns into the function
No_Prevalences=PLHIV(data_set['Prevalence_%'],data_set['NoPLHIV'])
```

```python
No_Prevalences
```

```python
#creating a column and populating with the number of people not living with HIV in the PLHIV column
data_set['PLHIV']=No_Prevalences
```

### What is the total NoPLHIV in all the cities (districts with the word “city” or “metro” in the name)? 

```python
#sum of districts with the word “city” or “metro” in the name
data_set['NoPLHIV'].where(data_set['District'].str.contains('City' or 'Metro')).sum() 
```
